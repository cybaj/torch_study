#### Pooling layers

##### AdaptiveAvgPool3d
  
Adaptive는 원하는 output size를 명시하는 방법으로 pooling 할 수 있어 편리하다. 

torch.nn.AdaptiveAvgPool3d(  
    output_size)
  
* output_size – the target output size of the image of the form D x H x W. Can be a tuple (D, H, W) or a single D for a cube D x D x D. D, H and W can be either a int, or None which means the size will be the same as that of the input.
  

Input: $(N, C, D_{in}, H_{in}, W_{in})$

Output: $(N, C, D_{out}, H_{out}, W_{out})$

```python
import torch.nn as nn
```

1. adaptiveavgpool3d

```python
help(nn.AdaptiveAvgPool3d)
```

```python
m1 = nn.AdaptiveAvgPool3d(5)
m2 = nn.AdaptiveAvgPool3d((3,5,1))
```

```python
inp = torch.randn(1,1,8,8,5)
```

```python
output1 = m1(inp)
print(output1.shape)
print(output1)
```

```python
output2 = m2(inp)
print(output2.shape)
print(output2)
```
