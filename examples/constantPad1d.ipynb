#### Padding layers

##### ConstantPad1d
  
상수값으로 padding 해준다.

torch.nn.ConstantPad1d(  
    padding,  
    value)  
  
* padding (int, tuple) – the size of the padding. If is int, uses the same padding in all boundaries. If a 2-tuple, uses padding_left, padding_right)
* value – constant
  

Input: $(N, C, W_{in})$

Output: $(N, C, W_{out})$

$$W_{out}=W_{in}+padding\_left+padding\_right$$

```python
import torch.nn as nn
```

```python
m = nn.ConstantPad1d(2, 7)
```
