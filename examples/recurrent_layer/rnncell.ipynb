#### Recurrent layers

##### RNN Cell
  
RNN Cell 하나

torch.nn.RNNCell(  
    input_size,  
    hidden_size,  
    bias=True, nonlinearity='tanh')
  
* **input_size** – The number of expected features in the input x
* **hidden_size** – The number of features in the hidden state h
  
  
* bias – If False, then the layer does not use bias weights b_ih and b_hh. Default: True
* nonlinearity – The non-linearity to use. Can be either 'tanh' or 'relu'. Default: 'tanh'

$$h=tanh(W_{ih}x_t+b_{ih}+W_{hh}h+b_{hh})$$

Input: $(N, H_{in})$

$H_{init}$: $(N, H_{out})$ initial hidden state for each element in the batch

Output: $(N, H_{out})$ next hidden state


```python
import torch.nn as nn
```

1. rnn cell

```python
help(nn.RNNCell)
```

```python
rnncell = nn.RNNCell(10, 20) # Cell 이므로 RNN stack 을 쌓지 않기 때문에, 해당 파라미터가 사라졌다. 
# No needs to specify the number of rnn layer, because it is just cell
# input feature map size : 10
# hidden state feature size : 20
```

```python
ipt = torch.randn(3, 10) # sequence 가 아니다.

# batch : 3
# input feature map size : 10

hx = torch.randn(3, 20) # 

# batch : 3
# 각 batch 별 hidden state
```

```python
output = rnncell(ipt, hx)
```

2. rnn cell 로 rnn layer 만들기 make a rnn layer by rnncells

```python
ipt = torch.randn(6, 3, 10) # input sequence
# seq_len : 6
# batch : 3
# input feature map size : 10
```

```python
h0 = torch.randn(3, 20) # 상태변수들의 초기값
# batch size : 3
# hidden state feature size : 20
```

```python
outputs = []
```

```python
for i in range(6):
    if i == 0 :
        hx = rnncell(ipt[i], h0)
    else :
        hx = rnncell(ipt[i], outputs[i-1])
    outputs.append(hx)
```
